# TASK 2.3 — MongoDB Queries

## Query 1: Quarter-final matches

```python
# Retrieve all matches where Stage is 'Quarter-finals'
for doc in collection.find({"Stage": "Quarter-finals"}).limit(5):
    print(doc)
```

In this query, the dataset is filtered to display only matches played in the quarter-final stage. This demonstrates how MongoDB can be used to retrieve specific records based on a given condition.

## Query 2: Projection (Selected Fields)

```python
# Display only selected fields: Year, Stage, HomeTeamName, AwayTeamName
result = collection.find({}, {
    "Year": 1,
    "Stage": 1,
    "HomeTeamName": 1,
    "AwayTeamName": 1,
    "_id": 0
})

for doc in result.limit(10):
    print(doc)
```

In this query, documents are filtered to display only quarter-final matches, and only selected fields such as Year, HomeTeamName, and AwayTeamName are retrieved. This demonstrates the use of projection in MongoDB to limit the returned data and improve query efficiency

## Query 3: Final matches

```python
# Retrieve only key details for final matches
for doc in collection.find(
    {"Stage": "Final"},
    {"Year": 1, "HomeTeamName": 1, "AwayTeamName": 1, "HomeTeamGoals": 1, "AwayTeamGoals": 1, "_id": 0}
).limit(10):
    print(doc)
```

## Query 4 (Delete operation)

```python
# Count how many will be deleted
collection.count_documents({"HomeTeamGoals": 0})
```

```python
# Delete all matches where HomeTeamGoals = 0
collection.delete_many({"HomeTeamGoals": 0})

print("Documents with HomeTeamGoals = 0 deleted")
```

```python
# Count how many will be deleted
collection.count_documents({"HomeTeamGoals": 0})
```

In this query, documents where the HomeTeamGoals value is equal to 0 are deleted from the collection. This demonstrates MongoDB’s ability to remove records based on a specified condition, supporting efficient data management and cleaning operations.

## Query 5 (Update operation)

```python
# Count how many documents will be updated
collection.count_documents({"RoundID": 308})
```

```python
# Update RoundID from 308 to 30800
collection.update_many(
    {"RoundID": 308},
    {"$set": {"RoundID": 30800}}
)

print("RoundID updated successfully")
```

```python
# Confirm update worked
collection.count_documents({"RoundID": 30800})
```

In this query, documents with a specific RoundID value are updated to a new value using the update operation. This demonstrates MongoDB’s capability to modify existing records efficiently without the need to recreate the dataset
